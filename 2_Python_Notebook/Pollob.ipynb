```python
import pandas as pd
import matplotlib.pyplot as plt

# Create segments
dfD["client_segment"] = pd.cut(
    dfD["client_tenure_months"],
    bins=[0, 36, 120, 210, 999],
    labels=["new", "mid-term", "established", "long-standing"]
)

# Count segments in correct order
segment_counts = dfD["client_segment"].value_counts().sort_index()

# Bar chart
segment_counts.plot(kind="bar")

plt.title("Client Segments by Tenure")
plt.xlabel("Client Segment")
plt.ylabel("Number of Clients")
plt.xticks(rotation=45)
plt.show()
```
